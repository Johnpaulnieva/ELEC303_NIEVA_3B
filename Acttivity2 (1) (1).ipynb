```python
accident['Accident Date'] = accident['Accident Date'].astype('str')
```

```python
accident['Accident Date'] = accident['Accident Date'].str.strip()
accident['Accident Date'] = accident['Accident Date'].astype('str')
accident['Accident Date'] = accident['Accident Date'].str.replace('/','-')
```

```python
accident['Accident Date'] = pd.to_datetime(accident['Accident Date'], dayfirst = True, errors = 'coerce')
```

<h1>Adjusting Data Types</h1>

```python
accident.dtypes
```

```python
accident['Year'] = accident['Accident Date'].dt.year
accident['Month'] = accident['Accident Date'].dt.month
accident['Day'] = accident['Accident Date'].dt.day
accident['DayOfWeek'] = accident['Accident Date'].dt.dayofweek
```

```python
accident.isnull().sum()
```

<h1>Data Visualization</h1>

```python
camden_accident = accident[accident['District Area'] == 'Camden']
```

```python
monthly_camdendata = camden_accident.groupby(['Year', 'Month']).size()

monthly_camdendata = monthly_camdendata.unstack(level=0)

monthly_camdendata.plot(
    marker='o', 
    figsize=(10, 5),
    title='Monthly Road Accidents in camden by Year (2019-2022)'
)
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
# plt.legend(title='Yearly Accident Comparison', bbox_to_anchor = (1.05,1), loc='upper left')
plt.grid(alpha=0.3)
plt.show()
```

<h3>Insight 26: 
Based on the provided graph, the number of monthly road accidents in Camden fluctuated between 2019 and 2022. The single most striking observation is the significant increase in accidents during November 2022, reaching the highest point recorded in the entire four-year period.</h3>

```python
severity_camden  = camden_accident['Accident_Severity'].value_counts()

plt.pie(severity_camden,
        labels = severity_camden.index,
        autopct = '%1.1f%%',
        startangle = 90)
plt.gcf().set_size_inches(10,7)
plt.title('Accident Severity Distribution')
plt.show()

severity_2022_camden = camden_accident[camden_accident['Year'] == 2022]
# severity_2019
lsc_2022 = severity_2022_camden['Accident_Severity'].value_counts()

plt.pie(lsc_2022,
        labels = lsc_2022.index,
        autopct = '%1.1f%%',
        startangle = 90)
plt.gcf().set_size_inches(10,7)
plt.title('Accident Severity Distribution')
plt.show()
```

<h3>Insight 27: In both charts, slight accidents make up the largest portion by a wide margin (85.3% in the top chart and 87.4% in the bottom chart), while fatal and serious accidents combined represent a small minority of the total accidents.</h3>

```python
# Accidents per year
accidents_per_year = camden_accident['Year'].value_counts().sort_index()
accidents_per_month = camden_accident['Month'].value_counts().sort_index()
accidents_per_day = camden_accident['DayOfWeek'].value_counts().sort_index()


#PER YEAR 
accidents_per_year.plot(kind='bar')
plt.title('Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.show()

month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Replace numerical months with their names
accidents_per_month.index = accidents_per_month.index.map(month_names)


accidents_per_month.plot(kind='bar')
plt.title('Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()


#accident per day
day_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

accidents_per_day.index = accidents_per_day.index.map(day_names)
accidents_per_day.plot(kind='bar')
print(accidents_per_day)
```

<h3>Insight 28: The number of accidents in the area increased each year from 2019 to 2022, with 2022 having the most accidents overall. Additionally, the most accidents generally happen in the middle of the year (May, June, and July) and on the weekends (Friday and Saturday).</h3>
